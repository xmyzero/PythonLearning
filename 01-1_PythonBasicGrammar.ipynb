#### 常见问题
1. 取变量名时，是否有注意事项？
> 程序员给变量取名是为了让代码可读性更高，方便记忆，而如何给变量有效命名，需要遵循以下规则：
    1. 第一个字符必须只能是字母或下划线，第一个字符不能出现数字或其他字符；
    2. 除第一个字符外，其他部分可以是字母或者下划线或者数字；
    3. 大小写敏感，比如name跟Name是不同的标识符。
    4. 避免与Python内置或使用模块已有的关键词、函数重复，如`print`等
> ***
    5. 变量名应即简短又具有描述性。如，name比n好，student_name比s_n好，name_length比length_of_persons_name好
    6. 谨慎使用小写字母l和大写字母O，避免与数字1和0混淆
2. 如何应对编程错误(bug)？

````python
# (is vs. ==)
# is 是检查两个变量是否指向同一个对象
# 但==是检查变量指向的对象是否是相同的值
a = [1, 2, 3, 4]  # 创建一个新的list对象——[1, 2, 3, 4]，赋值给变量a
b = a             # 将a再赋值给变量b
print(b is a)     # => True, a和b指向同一个对象
print(b == a)     # => True, a指向的对象和b指向的对象值相同
````

````python
b = [1, 2, 3, 4]  # 重新将list对象——[1, 2, 3, 4]，赋值给变量b
print(b is a)     # => False, a和b不是来自同一个对象
print(b == a)     # => True, a指向的对象和b指向的对象值相同
````

### if语句
[返回目录](#menu)

条件语句语法如下：
```
if 条件A（结果为布尔值，True或者False）:
    执行相关操作1（需要缩进）
    ......

elif 条件B（结果为布尔值，True或者False）:
    执行相关操作2
    ......
else:
    执行相关操作3
```

条件语句解释说明如下：
1. 若条件A为True，那么执行相关操作1，程序结束
2. 若条件A为False，那么判断条件B，若条件B为True，那么执行相关操作2，程序结束
3. 若条件A为False，那么判断条件B，若条件B为False，那么执行相关操作3，程序结束

条件语句注意事项如下：
1. 不同条件间要保持互斥
2. 不同条件的并集，尽可能等于全事件
3. 条件下的执行内容必须保持缩进

### for循环
[返回目录](#menu)

所有可迭代对象(iterable)都可以通过for循环遍历里面的所有内容，如list、tuple、dict、range等

for循环语法如下：  
```
for 变量 in 可迭代对象:
    循环执行语句
    ...
```

for循环解释说明如下：  
1. 每次循环提取可迭代对象内一个元素，赋值给in前面的变量
2. 可迭代对象内所有元素被遍历完后，循环结束
3. 循环次数和可迭代对象内元素个数有关

#### 练习12-1

````python
# 用for循环遍历整个列表
names = ['alice', 'david', 'caroline']
for name in names:
    print(name)
````

````python
# 在for循环中执行更多操作
names = ['alice', 'david', 'carolina']
for name in names:
    new_name = name.title()
    print(f'{new_name} is a teacher.\n')
#     print('He/She is a good teacher\n')
````

### while循环
[返回目录](#menu)

while循环语法如下：
```
while 条件:
    执行循环语句
    ...
```

while循环解释说明如下：
1. 判断条件，若条件为False，则循环结束

### try/except异常处理
[返回目录](#menu)

异常就是报错，报错会让我们的程序停止。  
在编程中，异常处理通常应用在我们认为不会出错，但是因某些原因导致的报错

异常处理的语法如下：
```
try:
    执行相关语句1
except Exception as e:
    执行相关语句2
else:
    执行相关语句3
finally:
    执行相关语句4
```

异常处理语法说明如下：
1. 先尝试执行相关语句1
2. 若在执行语句1的过程中报错，那么执行相关语句2
3. 若在执行语句1的过程中没有报错，那么执行相关语句3
4. 所有情况下都会执行相关语句4

#### 练习14-1

````python
# 用try/except块处理异常状况
try:
    i = 10
    assert i == 100
    # 用 "raise" 抛出异常
#     raise IndexError("This is an index error")
except IndexError as e:
    print(e)                 # 通常应该在这里处理错误
except (TypeError, NameError):
    pass                 # pass是无操作，可以同时处理不同类的错误
else:                    # try/except块的可选条件，必须跟在except后
    print("All good!")   # 只有当try运行完没有错误的时候，才会运行这一句
finally:                 # 在所有情况下都会执行
    print("We can clean up resources here")
````

````python
url = '新浪财经网的链接'
page = 1
flag = True
data = []
while flag:
    for i in range(3):
        try:
            print(f'访问{url}，第{page}页')
            assert page != error_page
        except:
            time.sleep(3)
        else:
            print(f'抓取第{page}页的数据')
            break
    else:
        break
    data.append(page)
    page += 1

print(f'保存数据{data}到本地')
````

### 函数
[返回目录](#menu)

函数定义的语法如下：
```
def 函数名(参数1, 参数2, ...):
    函数执行语句
    函数执行语句
    ...
    return 返回值
```

函数调用的语法如下：  
`函数名(传入参数1, 传入参数2)`

函数语法的说明如下：
1. 以def开头，表示define
2. 函数名(name)可以由自己取名，最好结合函数的意义
3. 函数可以包含任意个参数
       参数内容将会在函数内部执行语句时被调用；
       函数在定义时包含非默认参数(non-default argument)和默认参数(default argument)
       定义时的参数，统一被称之为形参。
4. 定义完函数名和参数后，需要跟上冒号:
5. 所有在函数内部执行的语句前都需包含缩进，内部执行语句逻辑与普通代码保持一致
6. return标志函数有返回值，函数运行完return语句之后，整个函数运行结束。函数可以没有return部分
7. 函数调用时传入的参数被称为实参
        是外部传给函数内部的实际变量
        会在函数调用过程中被使用

````python
# 如果一个函数在内部调用自身本身，这个函数就是递归函数
# 递归函数的优点是定义简单，逻辑清晰。理论上，所有的递归函数都可以写成循环的方式，但循环的逻辑不如递归清晰。
# 阶乘计算
def fact(n):
    if n==1:
        return 1
    return n * fact(n - 1)
````

````python
# 解决递归调用栈溢出的方法是通过尾递归优化
# 事实上尾递归和循环的效果是一样的，所以，把循环看成是一种特殊的尾递归函数也是可以的。
# 尾递归是指，在函数返回的时候，调用自身本身，并且，return语句不能包含表达式。

# 以阶乘为例
def fact_iter(num, product=1):
    if num == 1:
        return product
    return fact_iter(num - 1, num * product)
````

````python
# 传入函数
# 变量可以指向函数，函数的参数能接收变量，那么一个函数就可以接收另一个函数作为参数，这种函数就称之为高阶函数。
def add(x, y, f):
    return f(x) + f(y)

add(10, -21, lambda x: x>=10)
````

````python
# 用列表推导式可以简化映射和过滤。
# 列表推导式的返回值是另一个列表。
[add_10(i) for i in [1, 2, 3]]         # => [11, 12, 13]
````

````python
[add_10(x) for x in [3, 4, 5, 6, 7] if x > 5]  # => [6, 7]
````

````python
# 集合和字典也可以达到类似的效果
{x for x in 'abcddeef' if x not in 'abc'}  # => {'d', 'e', 'f'}
````
