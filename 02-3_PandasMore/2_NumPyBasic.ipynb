## 1.1 Ndarray对象
[返回目录](#menu)

NumPy 最重要的一个特点是其 N 维数组对象 ndarray(N-dimensional array object)，它是一系列同类型数据的集合，以 0 下标为开始进行集合中元素的索引。
- ndarray 对象是用于存放**同类型元素**的多维数组。
- ndarray 中的每个元素在内存中都有**相同存储大小的区域**。


```python
# 创建一个一维数组
arr = np.array([1, 2, 3])
arr
```

```python
# 创建一个多维数组
arr2 = np.array([[1, 2, 3], [4, 5, 5]])
arr2
```

```python
# 设置最小维度
a = np.array([1, 2, 3, 4, 5], ndmin=2) 
a
```

```python
# 设置内部元素的类型
a = np.array([1,  2,  3], dtype=complex)  
print(a)
a = np.array([1, 2, 3], dtype='f')
print(a)
```

## 1.4 数组的切片、索引与迭代
[返回目录](#menu)
### 1.4.1 切片
ndarray对象的内容可以通过索引或切片来访问和修改，与 Python 中 list 的切片操作一样

```python
[1,2,3]*3
```

```python
a = np.arange(10)*3
a
```

```python
# 一维数组的切片操作
print(a[0])
print(a[:8])  # 前闭后开
print(a[2::2])
```

```python
# 多维索引
print(arr2)
print('='*10)
print(arr2[:2, 1])
print(arr2[1:3, :])
print(arr2[1, 1])
```

```python
# 切片还可以用...
print(arr3)
print('='*10)
print(arr3.shape)
print(arr3[1, ...])  # 等价于arr3[1,:,:] 或arr3[1]
print(arr3[..., 2])  # 等价于arr3[:,:,2]
```

```python
arr4 = np.arange(120).reshape(2, 3, 4, 5)
arr4
```

```python
print(arr4[1, 2, ...])  # arr4[1,2,:,:]
print('*'*10)
print(arr4[..., 3])  # arr4[:, :, :, 3]
print('*'*10)
print(arr4[..., 3, :1])  # arr4[:, :, 3, :1]
```

### 1.4.2 索引

```python
arr2
```

```python
# 布尔索引
arr4[arr4>50]
```

```python
# 花式索引
# 花式索引根据索引数组的值作为目标数组的某个轴的下标来取值。
# 对于使用一维整型数组作为索引，如果目标是一维数组，那么索引的结果就是对应位置的元素；如果目标是二维数组，那么就是对应下标的行。
# 花式索引跟切片不一样，它总是将数据复制到新数组中。
x=np.arange(32).reshape((8,4))
x
```

```python
x[[4,2,1,7], :]
```

```python
x[[-4,-2,-1,-7]]
```

```python
df = pd.DataFrame(x, columns=list('abcd'))
df
```

```python
df.loc[:, ['b', 'a', 'd']]
```

## 1.5 数组的基础操作
[返回目录](#menu)


```python
arr
```

```python
print(arr.sum())
print(arr.min())
print(arr.max())
```

```python
print(arr.sum(axis=0))
print(arr.min(axis=1))
print(arr.cumsum(axis=1))
```

```python
# Universal Function(ufunc)
B = np.arange(3)
print(B)

print(np.exp(B))

print(np.sqrt(B))

print(np.sin(B))

C = np.array([2., -1., 4.])
print(np.add(B, C))
```

```python
# 广播(broadcast)
# 4x3 的二维数组与长为 3 的一维数组相加，等效于把数组 b 在二维上重复 4 次再运算
a = np.array([[ 0, 0, 0],
           [10,10,10],
           [20,20,20],
           [30,30,30]])
b = np.array([1,2,3])
print(a + b)
print(np.add(a, b))
```

```python
# numpy.maximum 
# numpy.minimum
# 比较不同序列的大小，并取最大值
s1 = pd.Series(np.random.randn(50))
s2 = pd.Series(np.random.randn(50))
np.maximum(s1, s2)
```

```python
# 如果是一个DataFrame中的两列，可以直接用max取最大值
df = pd.DataFrame({'s1': s1, 's2': s2})
df.max(axis=1)
```

```python
# np.where(condition, x, y)
# 满足条件(condition)，输出x，不满足输出y。
np.where(df['s1']>0, df['s2']+1, 0)
```
