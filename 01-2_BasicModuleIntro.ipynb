```python
for root, dirs, files in os.walk("./01_PythonGrammar", topdown=False):
    for name in files:
        print(os.path.join(root, name))
    for name in dirs:
        print(os.path.join(root, name))
```

## 3. `read()`, `readline()`, `readlines()`
[返回目录](#menu)

```python
import os
os.getcwd()
```

```python
# read
# 一次性读取
with open('mbox.txt') as f:
    mail = f.read()
    
print(mail, type(mail))
```

```python
# readline
# 逐行读取
with open('mbox.txt') as f:
    line = f.readline()
    print(line)
    line = f.readline()
    print(line)
    line = f.readline()
    print(line)
```

```python
bool('Received: from murder (mail.umich.edu [141.211.14.90])')
```

```python
with open('mbox.txt') as f:
    while True:
        line = f.readline()
        if line:
            print(line)
        else:
            print(line, type(line))
            break
```

```python
# readlines
# 将每一行内容存储为一个list
with open('mbox.txt') as f:
    lines = f.readlines()
    
lines
```
