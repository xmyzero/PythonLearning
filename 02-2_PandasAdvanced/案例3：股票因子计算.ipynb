```python
def count_fq(df):
    # 计算复权价格(在第一天买入1块钱的股票，最终能获得累积资金)
    df['涨跌幅'] = df['close'] / df['pre_close'] - 1
    df['复权因子'] = (1 + df['涨跌幅']).cumprod()
    df['close_fq'] = df['复权因子'] * (df.iloc[0]['close'] / df.iloc[0]['复权因子'])
    return df['close_fq']
```

```python
stock_df.groupby('ts_code', group_keys=False).apply(count_fq)
```

```python
stock_df['close_fq'] = stock_df.groupby('ts_code', group_keys=False).apply(count_fq)
stock_df
```

```python
# 按照复权收盘价计算因子
# alpha1因子公式（公式来自worldquant的101 formula alphas）：
# (rank(Ts_ArgMax(SignedPower(((returns < 0) ? stddev(returns, 20) : close), 2.), 5)) - 0.5)
# 公式说明：
# signedpower(x, a) = x的a次方
# ts_argmax(x, d) = 过去d天x序列最大值所对应的日期
# x?y:z = 如果x条件成立则返回y，否则返回z
# stddev(x, d) = 过去d天x序列的移动标准差
# rank(x) = x值的横截面排序（百分比）
# returns = daily close-to-close returns

def ts_argmax(df):
    # 计算: A = ((returns < 0) ? stddev(returns, 20) : close)
    rtn = df['close_fq'].pct_change()
    rolling_return_std = rtn.rolling(20, min_periods=1).std()
    df['rtn_factor'] = np.where(rtn<0, rolling_return_std, df['close_fq'])
    # 计算: B = Ts_ArgMax(SignedPower(A, 2.), 5)
    argmax = (df['rtn_factor']**2).rolling(5, min_periods=1).agg(pd.Series.idxmax)  # 获取过去五日最最大值对应的index
    df['ts_argmax'] = df.loc[argmax.values, 'trade_date'].values
    df.drop(columns=['rtn_factor'], inplace=True)
    return df
```

```python
stock_df = stock_df.groupby('ts_code', group_keys=False).apply(ts_argmax)
stock_df.head()
```

```python
# 计算: (rank(B)-0.5)
stock_df['alpha1'] = stock_df.groupby('trade_date').rank(pct=True)['ts_argmax']-0.5
stock_df.drop(columns=['ts_argmax'], inplace=True)
```
