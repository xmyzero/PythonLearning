```python
import pandas as pd
import numpy as np
```

```python
# 设置最新活动日期
activity_date = pd.to_datetime('20200529')

# 读入报名名单
bm = pd.read_csv('../datas/活动报名数据.csv', encoding='gbk', 
                 parse_dates=['创建时间'])  # 解析日期
bm  # 观察数据
```

```python
# 读入反馈名单
fk = pd.read_csv('../datas/活动反馈数据.csv', encoding='gbk', parse_dates=['创建时间'],
                true_values=['已完成'], false_values=['未完成'])
fk.head()
```

```python
# =已参加名单整理
bm['邮箱账号'] = bm['邮箱账号'].str.lower()  # 对邮箱账号进行统一处理，避免大小写敏感问题
# 找出过去已参加过的人员名单
old_list = bm.loc[bm['创建时间'] < activity_date, '邮箱账号'].unique()
old_list
```

```python
# =找出本次报名名单
new_list = bm.loc[bm['创建时间'] >= activity_date, '邮箱账号']
new_list
```

```python
# =从报名名单中剔除之前参加过的人员
qualify_new_list = new_list.loc[~new_list.isin(old_list)].unique()
qualify_new_list
```

```python
# =反馈名单数据整理
fk['账号'] = fk['账号'].str.lower()  # 对邮箱账号进行统一处理，避免大小写敏感问题

fk = fk.loc[fk['创建时间'] >= activity_date, :]  # 找出本次活动的反馈名单
fk
```

```python
fk[fk['创建时间'] <= pd.to_datetime('20200703 23:00:00')]
```

```python
# 找出完成任务的人员
fk = fk.loc[(fk['是否完成视频课程']) & (fk['是否完成大作业']), :]
fk
```

```python
# 找出符合条件(报名且过去未参加过活动)的领奖人名单
qualify_list = fk.loc[fk['账号'].isin(qualify_new_list), '账号'].unique()
qualify_list
```
